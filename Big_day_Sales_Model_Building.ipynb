# ``Interpretation`` 

#  Our target column is highly imbalanced. So we are considered ``F1-SCORE``.

``METRICS``


``ACCURACY`` Accuracy measures the proportion of correct predictions over the total number of predictions, and it is a good metric when the classes are balanced (i.e., each profit range has roughly the same number of instances).

``PRECISION``   Precision measures the proportion of true positive predictions over the total number of positive predictions, and it is a good metric when the cost of a false positive prediction is high.

``RECALL``    Recall measures the proportion of true positive predictions over the total number of actual positive instances, and it is a good metric when the cost of a false negative prediction is high. 

``F1-SCORE``  F1 score is the harmonic mean of precision and recall, and it is a good metric when you want to balance the trade-off between precision and recall.
